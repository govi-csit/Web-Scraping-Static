```python
#now combine above two logics to extract values for all indicators(stock market data)

def fetch_stock_market_data(tickerSymbol):
    
    #url = "https://finance.yahoo.com/quote/AAPL?p-AAPL"
    url = ("https://finance.yahoo.com/quote/" + tickerSymbol+
           "?p=" + tickerSymbol)

    #xtracting whole web page
    response = requests.get(url)
    htmlText = response.text
    
    #stock market data fields
    Indicators = ["Previous Close",
    "Open",
    "Bid"
    "Ask",
    "Day &#x27;s Range",
    "52 Week Range",
    "Volume",
    "Avg. Volume",
    "Market Cap",
    "Beta (5Y Monthly)",
    "PE Ratio (TTH)",
    "EPS (TTM)",
    "Earnings Date",
    "Forward Dividend &amp; Yield",
    "Ex-Dividend Date",
    "ly Target Est"
    ]

    data = {}

    #keeping indicators in a list for which our Logic broken; need to handle seperately
    missingIndicators = [
        'Day&#x27;s Range',
        '52 Week Range',
        'Forward Dividend &amp; Yield'
        ]

    #for all the indicators
    for label in Indicators:

        try:
            #splitting text wrt indicator
            SplitList = htmlText.split(label)
        
            #check indicator type and split accordingly
            if(label in missingIndicators):
                afterFirstSplit = SplitList[1].split("\">")[1]
                afterSecondSplit = afterFirstSplit.split("</td>")
            else:
                afterFirstSplit = SplitList[1].split("\">")[2]
                afterSecondSplit = afterFirstSplit.split("</span>")
        
            #capturing data value
            dataValue = afterSecondSplit[0]
            data.update({label : dataValue})
    
        except :
            data.update({label : 'Exception Raise'})

    return data

#calling the function for validation
stock_data = fetch_stock_market_data('AAPL')
stock_data
```

Approach:

```python
wikiurl = "https://en.wikipedia.org/wiki/List_of_S&P_500_companies"
wikiResponse = requests.get(wikiurl)

print(wikiResponse.text)
```

```python
#cutting off the page at Last row of the table
wickiFirstParse = wikiResponse.text.split("0001555280")[0]

print(wickiFirstParse[5000:10000])
```

```python
#cutting off the page at starting row of the table

wickiDataTable = wickiFirstParse.split("S&amp;P 500 component stocks")[3]
print(wickiDataTable)
```

```python
#now we have table data in wikiData Table. Lets split it by hyperLink as the ticker symbols contains hyper links and see
    # any pattern can be identifiable to extract all ticker symbols

print(wickiDataTable.split("href=")[0], "\n")
print(wickiDataTable.split("href=")[1], "\n")
print(wickiDataTable.split("href=")[2], "\n")
print(wickiDataTable.split("href=")[3], "\n")
print(wickiDataTable.split("href=")[4], "\n")

print (wickiDataTable.split("href=")[5], "\n")
```

<h4>observation : </h4>every row contains 4 hyper links and the first symbol starts at 5th index; It means next symbol might
be at 5+4 = 9th index, and the next symbol at 9+4 = 13th index, and the same pattern may continue

```python
#verifying pattern

print(wickiDataTable.split("href=")[5], "\n")
print(wickiDataTable.split("href")[9], "\n")
print(wickiDataTable.split("href=")[13], "\n")
print(wickiDataTable.split("href=")[17], "\n")
```

```python
#split further to fetch ticker symbol

print(wickiDataTable.split("href=")[5].split(">"), "\n")
print(wickiDataTable.split("href=")[5].split(">")[1].split("</")[0])
```

```python
#Lets use the above identified pattern to see ticker symbol's hyper Links

for i in range(5, len(wickiDataTable.split("href=")), 4):
    print(wickiDataTable.split("href=")[i], "\n")
```

```python
# extract the ticker symbols using above mentioned pattern

data = {"Company" : []}
#splitting table data using hyper links
hyperLinkSplitwicki = wickiDataTable.split("href=")

#iterting over all hyper Links and capturing ticker symbols
for index in range (5, len(hyperLinkSplitwicki), 4):
    temp_data = hyperLinkSplitwicki[index].split(">")[1].split("</")[0]
    data["Company"].append(temp_data)


print(data["Company"])
```

<h3>Differnt pattern / Logic : </h3>

when you see the web page source code, we can observe that all the ticker symbols contains either of the bel
1. https://www.nyse.com/quote/XNYS
2. http://www.nasdaq.com/symbol/         
we need to use these links to identify ticker symbols

<img src="Images/ticker_symbol_hyperlinks.PNG">

```python
data = {"Company" : []}

#spiltting table data using href(hyper Link)
hyperLinkSplitWiki = wickiDataTable.split("href=")

#iterating over all hyper links and check whether Link has either one of two the hyper Links (key workds)
for index in range (0, len(hyperLinkSplitWiki)):
    if( ("nyse" in hyperLinkSplitWiki[index] and "quote" in hyperLinkSplitWiki[index]) or
    ("nasdaq" in hyperLinkSplitWiki[index] and "symbol" in hyperLinkSplitWiki[index])
    ):
        #capturing ticker symbol
        temp_data = hyperLinkSplitWiki[index].split(">")[1].split("</")[0]
        data["Company"].append(temp_data)

print(data["Company"])
print("total number of symbols : ", len(data["Company"]))
```
